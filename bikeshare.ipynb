```python
def clean_date(df):
#     split the start time and end time column
    df[["start_date", "start_time"]] = df["Start Time"].str.split(" ", expand = True)
    df[["end_date", "end_time"]] = df["End Time"].str.split(" ", expand = True)
#     Drop the start time and end time column
    df = df.drop(["Start Time", "End Time"], axis = 1)
#     Convert start time, start date and end date column to datetime format
    df["start_date"] = pd.to_datetime(df["start_date"])
    df["end_date"] = pd.to_datetime(df["end_date"])
    df["start_time"] = pd.to_datetime(df["start_time"])
#     Get the day, month and hour from the start ate and time column repectively
    df["DayOfWeek"] = df["start_date"].dt.weekday
    df["start_month"] = df["start_date"].dt.month
    df["Hour"] = df["start_time"].dt.hour
    return df


clean_date = clean_date(file)
clean_date.head()
```

```python
def time_stats(df):
    # Display the most common month
    most_common_month = df["start_month"].value_counts().idxmax()
    month = dict(enumerate(calendar.month_name))
    
    # Display the most common day of the week
    most_common_day = df["DayOfWeek"].value_counts().idxmax()
    days = dict(enumerate(calendar.day_name))
    
    # Display the most common hour
    most_common_hour = df["Hour"].value_counts().idxmax()
    
    
    print("---Most common month---")
    print(f"{month[most_common_month]}\n")
    
    print("---Most common day of the week---")
    print(f"{days[most_common_day]}\n")
    
    print("---Most common hour---")
    print(most_common_hour)

time_stats = time_stats(clean_date)
```

```python
def station_stats(df):
    # Most common start station
    most_common_start_station = df["Start Station"].value_counts().idxmax()
    
    # Most common end station
    most_common_end_station = df["End Station"].value_counts().idxmax()
    
    # Most common combination of start and end station
    most_common_station = df[["Start Station", "End Station"]].value_counts().idxmax()
    
    print("---Most common start station---")
    print(f"{most_common_start_station}\n")
    
    print("---Most common end station---")
    print(f"{most_common_end_station}\n")
    
    print("---Most common combination of start and end station---")
    print(most_common_station)
    
station_stats(clean_date)
```

```python
# total and mean travel time
def travel_stats(df):
    # Display the total travel time
    total_travel_time = df["Trip Duration"].sum()
    
    # Display the mean travel time
    average_travel_time = df["Trip Duration"].mean()
    
    print("---Total Travel Time---")
    print(f"{total_travel_time}\n")
    
    print("---Average Travel Time---")
    print(round(average_travel_time, 2))
    
travel_stats(clean_date)
```

```python
def user_stats(df):
    # Display counts of user types
    if "Gender" and "Birth Year" in df.columns:
        users_df = df["User Type"].value_counts().reset_index()
        users_df.columns = ["Type", "Count"]
        
        # Display counts of gender
        gender_df = df["Gender"].value_counts().reset_index()
        gender_df.columns = ["Gender", "Count"]

        # Display earliest year of birth
        earliest_birth_year = df["Birth Year"].min()

        # Display most recent year of birth
        recent_birth_year = df["Birth Year"].max()

        # Display most common year of birth
        most_common_birth_year = df["Birth Year"].value_counts().idxmax()
        
        
        print("---User type count---")
        print(f"{users_df}\n")
        
        print("---Gender count---")
        print(f"{gender_df}\n")
        
        print("---Earliest year of birth---")
        print(f"{earliest_birth_year}\n")
        
        print("---Most recent year of birth---")
        print(f"{recent_birth_year}\n")
        
        print("---Most common year of birth---")
        print(most_common_birth_year)
    else:
        count_user_info = df["User Type"].value_counts()
        print("----count user type----")
        print(count_user_info)

        
user_stats(clean_date)
```
